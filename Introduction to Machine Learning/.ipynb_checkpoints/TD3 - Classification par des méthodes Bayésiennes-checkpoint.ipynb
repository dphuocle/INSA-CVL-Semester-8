```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi
x = np.linspace(-pi, 3*pi, 300)
pxc1 = np.exp(-((x-2)**2)/2)/((2*pi)**(0.5))
pxc2 = np.exp(-((x-4)**2)/8)/((4*2*pi)**(0.5))
pc1 = pc2 = 0.5
px = pxc1 * pc1 + pxc2 * pc2
pc1x = (pxc1 * pc1) / px
pc2x = (pxc2 * pc2) / px

plt.plot(x,pxc1,x,pxc2,'r')
```
